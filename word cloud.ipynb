```python
def func():
    import pandas as pd
    import matplotlib.pyplot as plt 
    from wordcloud import WordCloud 
    from nltk.tokenize import word_tokenize

    fopen1=open("data_mining.doc","r")
    fopen2=open("web_mining.doc","r")
    w1=fopen1.read()
    w2=fopen2.read()

    token1=word_tokenize(w1)
    token2=word_tokenize(w2)

    d1={}
    for k in token1:
       if k in d1:
          d1[k]+=1
       else:
          d1[k]=1

    d2={}
    for k in token2:
       if k in d2:
          d2[k]+=1
       else:
          d2[k]=1


    wordcloud = WordCloud(width=900,height=500, max_words=1628,relative_scaling=0.1,normalize_plurals=False).generate_from_frequencies(d1)
    wordcloud1 = WordCloud(width=900,height=500, max_words=1628,relative_scaling=0.1,normalize_plurals=False).generate_from_frequencies(d2)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

    plt.imshow(wordcloud1, interpolation='bilinear')
    plt.axis("off")
    return(plt.show())
```
