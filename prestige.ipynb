```python
import nltk
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize

fopen1=open("data.txt","r")
fopen2=open("data1.doc","w")

for line in fopen1:
    new_line = line.replace('\t', ' ')
    fopen2.write(new_line)
fopen1.close()
fopen2.close()

fopen1=open("data1.doc","r")
fopen2=open("data2.doc","w")
fopen3=open("data3.doc","w")

for line in fopen1:
    new_line1 = line.split(" ")
    fopen2.write(new_line1[0])
    fopen2.write("\n")
    fopen3.write(new_line1[1])
    fopen3.write("\n")

f=open("data2.doc","r")
t=open("data3.doc","r")
fromn=f.read()
w1=fromn.split()
ton=t.read()
w2=ton.split()
d=nx.DiGraph(Directed=True)
for x in range(102747):
    d.add_edge(w1[x], w2[x])
n_nodes=d.number_of_nodes()
degree_prestige = dict((v,len(d.in_edges(v))/(n_nodes-1)) for v in d.nodes())
print("Degree Prestige:\n")
print ("Node \t Degree Prestige")
for i in degree_prestige:
    print(i, " \t ", degree_prestige[i])
    
prox=[]
for dest in d.nodes:
    s=0
    dist=0
    I=0
    dist=nx.single_source_shortest_path_length(d,dest)
    for i in dist:
        s+=dist[i]
    I=len(dist)
    if I==0 or s==0:
        prox.append([dest, 0])
    else:
        prox.append([dest, (I/(n_nodes-1))/(s/I)])
print("\nProximity Prestige:\n")
print ("Node \t Proximity Prestige")
for i in prox:
    print(str(i[0]) + " \t " + str(i[1]))


prominance = np.random.randint(1, n_nodes-1, size=n_nodes)
rank_prestige = np.zeros([n_nodes], dtype = int)
path_matrix = np.zeros([n_nodes, n_nodes], dtype = int)
i = 0
j = 0
for src in d.nodes:
    for dest in d.nodes:
        if d.has_edge(dest, src):
            path_matrix[i][j] = 1
            j = j+1
    j = 0
    i = i+1
for i in range(0,n_nodes):
    pr_i = 0
    for j in range(n_nodes):
        pr_i = pr_i + path_matrix[i][j] * prominance[j]
    rank_prestige[i] = pr_i

print("Rank Prestige:\n")
print ("Node \t Rank Prestige")
for i in range(0,n_nodes):
    print(i+1, " \t ", rank_prestige[i])
```
