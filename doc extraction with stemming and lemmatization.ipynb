```python
import re
import nltk
from bs4 import BeautifulSoup
from urllib import request
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer 
import pandas as pd

url="https://en.wikipedia.org/wiki/Web_mining"
html=request.urlopen(url).read().decode('utf8')
soup=BeautifulSoup(html, 'html.parser')
l=[]
term=[]
for titles in soup.find_all(['h1', 'h2','h3','h4','h5','h6', 'p']):
    l=word_tokenize(titles.text)
    term=term+l

print(term)
```

```python
stop_words=list(set(stopwords.words('english')))

filtered=[]
for w in term:
    if w not in stop_words:
        filtered.append(w)
print("Total number of terms after stopword removal: ", len(filtered))

count={}
for element in filtered:
   if element in count:
      count[element]+=1
   else:
      count[element]=1
print("Term frequency after stopword removal")
print(count)
```

```python
stop_words_add=[',', '.', '-', '(', ')', '[', ']', ':', ';', '\'']
for words in stop_words_add:
    stop_words.extend(words)

final_term=[]
for w in filtered:
    if w not in stop_words:
        final_term.append(w)
print("Terms after removal of special characters")
print(final_term)
```

```python
stemmer=SnowballStemmer(language='english')
lemmatizer=WordNetLemmatizer()

stemming=[]
lemmatising=[]
for w in final_term:
    stemming.append(stemmer.stem(w))
    lemmatising.append(lemmatizer.lemmatize(w))

df1=pd.DataFrame(zip(final_term, stemming, lemmatising))
df1.columns=['Original Term', 'Stemmed Term', 'Lemmatized Term']
print(df1)
```

```python
u_stemming = []
u_lemmatising = []

for w in stemming:
    if w not in u_stemming:
        u_stemming.append(w)
for w in lemmatising:
    if w not in u_lemmatising:
        u_lemmatising.append(w)
print("Number of words after stemming")
print(len(u_stemming))
print("Number of words after lemmatising")
print(len(u_lemmatising))
```

```python
stop_words=list(set(stopwords.words('english')))
sentence_list=[]
l=[]
for titles in soup.find_all(['h1', 'h2','h3','h4','h5','h6', 'p']):
    l=(titles.text).split('.')
    sentence_list+=l
    
for sentence in sentence_list:
        print(sentence)
        words=[]
        stopword=[]
        words=word_tokenize(sentence)
        for word in words:
            if word in stop_words:
                stopword.append(word)
        a=pd.DataFrame(nltk.pos_tag(stopword))
        if a.empty:
            print("No stop words in this sentence")
        else:
            print(a)
```
